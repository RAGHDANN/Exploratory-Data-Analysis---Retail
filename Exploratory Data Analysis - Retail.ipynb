## Exploratory Data Analysis




## Which Category has heighst sales and profit?

```python
df.groupby(['Category'])[['Profit','Sales']].sum()
```

#### Conculosion 1: Technology(Category) has heighst sales and profit

##  Which Sub-Categories have high profit and profit loss?

```python
print(df.groupby('Sub-Category').Profit.sum().sort_values())
```

#### Conculosion 2: Top three Sub-Categories in profit : Copiers, Phones, Accessories ,and Sub-Categories that have profit loss: Tables, Bookcases, Supplies

```python
dc=df.groupby(['Sub-Category']).sum()[['Profit','Sales']].sort_values(['Profit'])
dc.plot(kind='bar',figsize=(16,10),color = ['Black','Blue']);
plt.title('Sum of profit & sales For Sub-Category');
```

```python
df.groupby(['State','Region','Sub-Category']).sum().idxmax()
```

#### Conculosion 3: The State, Region, and Sub-Category together that have maximum sales:(California, West, Phones) and maximum profit : (New York, East, Machines)

```python
df.groupby(['State','Region','Sub-Category']).sum().idxmin()
```

#### Conculosion 4: The State, Region, and Sub-Category together that have minimum sales:(Wisconsin, Central, Fasteners) and minimum profit : (Texas, Central, Binders)

```python
df.query('Profit<0.0000').groupby('State').count().Profit.sort_values().plot(kind='barh',figsize=(13,9),color = ['Pink','Pink','Pink','Pink','Pink','Pink','Pink','Pink','Pink','Pink','Pink','Pink','Pink','Pink','Pink','Pink','Pink','Pink','Pink','Pink','Pink','Pink','purple']);
plt.title('no.profit loss in states');
```

#### Conculosion 5: As we see at the bar chart, we found that Texas has the highest amount of profit loss

```python
ds=df.groupby(['State']).mean()[['Profit','Sales']].sort_values('Profit')
ds.plot(kind='bar',figsize=(16,10),color = ['Red','orange']);
plt.title('Mean of profit & sales in states');
```
